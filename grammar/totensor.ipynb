# 均值与方差

## 均值

### 遍历

```python
train_mean = 0
for x,y in server_train_data:
    # train_mean += x.mean()
    train_mean += x.sum()
train_mean = train_mean / len(server_train_data) / 28 / 28
```

```python
train_mean
```

```python
train_mean = 0
for x,y in server_train_data:
    train_mean += x.mean()
train_mean = train_mean / len(server_train_data)
```

```python
train_mean
```

### 除以255

```python
(server_train_data.data / 255).mean()
```

## 方差

### 遍历

```python
train_std = 0
for x,y in server_train_data:
    train_std += torch.pow(x-train_mean,2).sum()

torch.sqrt(train_std / len(server_train_data) / 28 / 28)
```

```python
train_std = 0
for x,y in server_train_data:
    train_std += x.var()

torch.sqrt(train_std / len(server_train_data))
```

如果直接用该对象求均值，需要遍历、累加，但不需要除以255，因为本身范围就是[0,1]

如果用data属性求均值，直接可以用mean()函数，但需要除以255

```python
(server_train_data.data / 255).std()
```

# 标准正态分布(有normalize)

获得上面这两个数后，可以知道MNIST符合(0.1307,0.3081)的正态分布，然后可以通过$\frac{x-u}{\sigma}$转为标准正态分布。

```python
transform = torchvision.transforms.Compose([
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize((0.1307,),(0.3081,)),
    ])
```

```python
train_dataset = torchvision.datasets.MNIST(
    data_dir,
    True,
    transform,
    download=True
    )
```

```python
train_dataset[0][0]
```

再次计算均值与方差, 应该是(0,1)

```python
train_mean = 0
for x,y in train_dataset:
    train_mean += x.mean()

train_mean = train_mean / len(train_dataset)
```

```python
train_mean
```

```python
train_std = 0
for x,y in train_dataset:
    train_std += torch.pow((x-train_mean),2).sum()

train_std = torch.sqrt(train_std/len(train_dataset)/28/28)
```
