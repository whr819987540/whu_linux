```python
import matplotlib.pyplot as plt
import numpy as np
```

## 自己构建正态分布函数

```python
# 正态分布函数
def normal(x: np.ndarray, u, sigma):
    return 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-0.5/sigma**2*(x-u)**2)
```

```python
para = [0, 3.6]
x = np.arange(-10, 10, 0.1)
y = normal(x, para[0], para[1],)
plt.ylabel("p(x)")
plt.xlabel("x")
```

```python
plt.plot(x, y, label="normal distribution")
plt.show()
```
